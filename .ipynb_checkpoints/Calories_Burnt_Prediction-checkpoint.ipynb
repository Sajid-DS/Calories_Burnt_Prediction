## Importing Important Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset 

```python
cal = pd.read_csv('calories.csv')
exc = pd.read_csv('exercise.csv')
```

## Performing  EDA

```python
#printing top rows in dataset
cal.head()
```

```python
exc.head()
```

```python
#Merging two datasets
data = pd.concat([exc,cal['Calories']], axis=1)
```

```python
data.head()
```

```python
#showing some stats about data
data.describe()
```

```python
#shape of data (rows,cols)
data.shape
```

## Data Visualization

```python
sns.set()
```

```python
#showing  numbers of male and females in dataset through countplot
sns.countplot(data['Gender'])
```

```python
  
sns.pairplot(data,kind='hist')
```

```python
corel = data.corr()
```

```python
#Plotting HeatMap
plt.figure(figsize=(10,10))
sns.heatmap(corel, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':9},cmap='Reds')
```

```python
#Converting Categorical Data to Numerical Data 
data.replace({'Gender':{'male':0,'female':1}},inplace=True)
```

```python
data.head()
```

```python
#Allocating features and target columns respectively to X & y variables
X = data.drop(columns=['User_ID','Calories'],axis=1)
y = data['Calories']
```

```python
print(X)
```

```python
y
```

```python
#Applying train_test_split method to divide data into train set & test set 
X_train, X_test,y_train, y_test = train_test_split(X.values,y.values,test_size=0.2)
```

```python
print(X_test)
```

### Applying machine learning Algorithm

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
pred_lin = lin_reg.predict(X_test)
```

```python
print(mean_absolute_error(y_test,pred_lin))
```

```python
#Checking Performance 
print('Performance in training set-->',lin_reg.score(X_train,y_train))
print('Performance in testing set -->',lin_reg.score(X_test,y_test))
```

```python
#creating XGBootRegressor Model
model = XGBRegressor()
```

```python
#Fitting Data to model
model.fit(X_train,y_train)
```

```python
#Predicting data and store this prediction_result into variable
prediction = model.predict(X_test)
```

```python
#Checking Performance 
print('Performance in training set-->',model.score(X_train,y_train))
print('Performance in testing set -->',model.score(X_test,y_test))
```

```python
print(mean_absolute_error(y_test,prediction))
```

```python
#Creating Random Forest Regressor model with default parameters
model_2 = RandomForestRegressor()
```

```python
model_2.fit(X_train,y_train)
```

```python
pred_Rf = model_2.predict(X_test)
```

```python
#Checking Performance 
print('Performance in training set-->',model_2.score(X_train,y_train))
print('Performance in testing set -->',model_2.score(X_test,y_test))
```

```python
print(mean_absolute_error(y_test,pred_Rf))
```

```python
#XGBoost is performing well in this problem 
sns.displot(y_test-prediction)
```
